### Exploratory Data Analysis
**This notebook will explore data sets of crime and socioeconomic data for the City of Chicago between the years of 2008 - 2011.**

```python
% matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime
import seaborn as sb
import matplotlib.ticker as mtick
import itertools
```

**Import the crimes and socioeconomic data sets into pandas dataframes.**

```python
crimes = pd.read_csv('Chicago_Crimes_2008_to_2011.csv', warn_bad_lines = True, error_bad_lines = False)
```

```python
socioeconomic = pd.read_csv('Chicago_socioeconomic_2008_2012_withpop.csv')
```

**Begin exploring and cleaning the crimes data set.**

```python
crimes.head(2)
```

```python
list(crimes.columns.values)
```

```python
del crimes['Unnamed: 0']
```

```python
crimes.columns = ['ID', 'case_number', 'date', 'block', 'IUCR', 'primary_type', 'description', 'location_description',
                  'arrest', 'domestic', 'beat', 'district', 'ward', 'community_area', 'FBI_code', 'x_coordinate',
                  'y_coordinate', 'year', 'updated_on', 'latitude', 'longitude', 'location']
```

```python
crimes.head(2)
```

```python
crimes.isnull().sum()
```

```python
crimes = crimes[np.isfinite(crimes['community_area'])]
```

```python
crimes.isnull().sum()
```

```python
crimes.date.dtype
```

```python
crimes.date = pd.to_datetime(crimes.date, format = '%m/%d/%Y %I:%M:%S %p')
```

```python
crimes.date.dtype
```

```python
crimes.head(2)
```

**Begin exploring and cleaning the socioeconomic data set.**

```python
socioeconomic.head(3)
```

```python
socioeconomic.columns = ['community_area', 'community_area_name', 'percent_housing_crowded',
                         'percent_households_below_poverty', 'percent_over16_unemployed',
                         'percent_over25_without_diploma', 'percent_under18_over64', 'income_per_capita',
                         'hardship_index', '2010_population']
```

```python
socioeconomic['2010_population'] = socioeconomic['2010_population'].str.replace(',','')
```

```python
socioeconomic['2010_population'] = pd.to_numeric(socioeconomic['2010_population'])
```

```python
socioeconomic.head(3)
```

```python
socioeconomic.tail(3)
```

```python
socioeconomic.drop(socioeconomic.index[-1:], inplace = True)
```

**Add community area name to crimes dataframe.**

```python
crimes = crimes.merge(socioeconomic[['community_area','community_area_name']], on = 'community_area', how = 'left')
```

**Plot crimes and arrests over time to see if there are any clear patterns.**

```python
arrests = crimes[['date', 'arrest']]
arrests = arrests.set_index('date')
crimes_by_month = arrests.resample('M').count().reset_index()
crimes_by_month.columns = ['date', 'crimes']
arrests_by_month = arrests.resample('M').sum().reset_index()
arrests_by_month.columns = ['date', 'arrests']
```

```python
fig, ax = plt.subplots(figsize = (12,5))
ax.plot(crimes_by_month['date'], crimes_by_month['crimes'], label = 'crimes')
ax.plot(arrests_by_month['date'], arrests_by_month['arrests'], label = 'arrests')
ax.legend()
ax.set_xlabel('Date')
ax.set_ylabel('Total Incidents')
ax.set_title('Crimes and Arrests in Chicago by Month')

fig.savefig('crime_by_month.png', transparent = False, dpi = 80, bbox_inches = 'tight')
```

**Seperate violent crime for future analysis.**

```python
crimes.primary_type.unique()
```

```python
# per http://gis.chicagopolice.org/clearmap_crime_sums/crime_types.html
violent_crimes = ['HOMICIDE', 'BATTERY', 'CRIM SEXUAL ASSAULT', 'ASSAULT', 'ROBBERY']
```

```python
def violence(row):
    if row['primary_type'] in violent_crimes:
        value = True
    else:
        value = False
    return value
```

```python
crimes['violent'] = crimes.apply(lambda x: violence(x), 1)
```

**Start to look at crime on the community level.  First look at yearly crime rate by community.**

```python
arrests_by_community = crimes[['arrest', 'community_area_name', 'violent']]
crimes_per_community = arrests_by_community.groupby(arrests_by_community['community_area_name']).count().reset_index()
crimes_per_community.head()
del crimes_per_community['violent']
crimes_per_community.columns = ['community_area_name', 'crimes']
arrests_per_community = arrests_by_community.groupby(arrests_by_community['community_area_name']).sum().reset_index()
arrests_per_community.columns = ['community_area_name', 'arrests', 'violent_crimes']
arrest_percent_per_community_total = crimes_per_community.merge(arrests_per_community[['arrests',
                                    'violent_crimes']], left_index = True, right_index = True, how = 'left')
```

```python
arrest_percent_per_community_total.head()
```

```python
crime_per_1000_total = arrest_percent_per_community_total.merge(socioeconomic[['community_area_name', 
                       '2010_population', 'income_per_capita', 'percent_over25_without_diploma', 
                       'percent_households_below_poverty', 'percent_over16_unemployed']], 
                        on = 'community_area_name', how = 'left')
```

```python
crime_per_1000_total['mean_yearly_crime'] = (crime_per_1000_total['crimes'] / 4)
crime_per_1000_total['mean_yearly_violent_crime'] = (crime_per_1000_total['violent_crimes'] / 4)
crime_per_1000_total['mean_yearly_crime_rate_per1000'] = crime_per_1000_total['mean_yearly_crime'] / (crime_per_1000_total['2010_population']/1000)
crime_per_1000_total['mean_yearly_violent_rate_per1000'] = crime_per_1000_total['mean_yearly_violent_crime'] / (crime_per_1000_total['2010_population']/1000)
```

```python
crime_per_1000_total.head()
```

```python
community_crime_sort = crime_per_1000_total.sort_values(by = 'mean_yearly_crime_rate_per1000', ascending = False)
community_crime_sort.set_index(['community_area_name'], inplace = True)
fig = plt.figure(figsize = (12,5))
ax = community_crime_sort['mean_yearly_crime_rate_per1000'].plot(kind = 'bar')
ax.set_xlabel('Community')
ax.set_ylabel('Yearly Crimes Per 1,000 Residents')
ax.set_title('Crime Rates per 1,000 Residents in each Community')

fig.savefig('crime_rate_per_community.png', transparent = False, dpi = 80, bbox_inches = 'tight')
```

**Examine the population of each community.**

```python
population_sort = socioeconomic.sort_values(by = '2010_population', ascending = False)
population_sort.set_index(['community_area_name'], inplace = True)
fig = plt.figure(figsize = (12,5))
ax = population_sort['2010_population'].plot(kind = 'bar')
ax.set_xlabel('Community')
ax.set_ylabel('Population in 2010')
ax.set_title('Population in each Community')

fig.savefig('population_per_community.png', transparent = False, dpi = 80, bbox_inches = 'tight')
```

**Examine crimes and arrests by month in each community.**

```python
arrests_by_community = crimes[['date', 'arrest', 'community_area_name', 'violent']]
arrests_by_community = arrests_by_community.set_index('date')
crimes_by_month_per_community = arrests_by_community.groupby([arrests_by_community['community_area_name'],
                                                              pd.TimeGrouper(freq = '1M')]).count().reset_index()
del crimes_by_month_per_community['violent']
crimes_by_month_per_community.columns = ['community_area_name', 'date', 'crimes',]
arrests_by_month_per_community = arrests_by_community.groupby([arrests_by_community['community_area_name'],
                                                               pd.TimeGrouper(freq = '1M')]).sum().reset_index()
arrests_by_month_per_community.columns = ['community_area_name', 'date', 'arrests', 'violent_crimes']
arrest_percent_per_community_month = crimes_by_month_per_community.merge(arrests_by_month_per_community[['arrests', 
                                            'violent_crimes']], left_index = True, right_index = True, how = 'left')
```

```python
arrest_percent_per_community_month.head()
```

```python
arrest_percent_per_community_month.tail()
```

**Look at crimes by month by community for communities in top 10 and bottom 10 for crime.**

```python
crime_sort = crime_per_1000_total.sort_values(by = 'mean_yearly_crime_rate_per1000', ascending = False)
top_10_crime = list(crime_sort['community_area_name'][:10])
bottom_10_crime = list(crime_sort['community_area_name'][-10:])
```

```python
fig, ax = plt.subplots(figsize = (12,10))
alpha_level = 1

for community in top_10_crime:
    ax.plot(arrest_percent_per_community_month['date'][arrest_percent_per_community_month.community_area_name == community], 
            arrest_percent_per_community_month['crimes'][arrest_percent_per_community_month.community_area_name == community],
            alpha = alpha_level, label = community+'(T)', color = 'r')
    alpha_level -= .1

for community in bottom_10_crime:
    alpha_level += .1
    ax.plot(arrest_percent_per_community_month['date'][arrest_percent_per_community_month.community_area_name == community], 
            arrest_percent_per_community_month['crimes'][arrest_percent_per_community_month.community_area_name == community],
            alpha = alpha_level, label = community+'(B)', color = 'b')

ax.set_xlabel('Date')
ax.set_ylabel('Crimes per Month')
ax.set_title('Crimes Per Month for Communities with Top 10 (T) and Bottom 10 (B) Crime Rates')
ax.legend()

fig.savefig('crimes_per_month_by_community.png', transparent = False, dpi = 80, bbox_inches = 'tight')
```

**Determine crimes per 1,000 residents per month per community.**

```python
crime_per_1000 = arrest_percent_per_community_month.merge(socioeconomic[['community_area_name', '2010_population']], 
                                                    on = 'community_area_name', how = 'left')
```

```python
crime_per_1000['c_per_1000residents'] = crime_per_1000['crimes'] / (crime_per_1000['2010_population']/1000)
crime_per_1000['v_per_1000residents'] = crime_per_1000['violent_crimes'] / (crime_per_1000['2010_population']/1000)
```

```python
crime_per_1000.head()
```

```python
fig, ax = plt.subplots(figsize = (12,10))
alpha_level = 1

for community in top_10_crime:
    ax.plot(crime_per_1000['date'][crime_per_1000.community_area_name == community], 
            crime_per_1000['c_per_1000residents'][crime_per_1000.community_area_name == community],
            alpha = alpha_level, label = community+'(T)', color = 'r')
    alpha_level -= .1

for community in bottom_10_crime:
    alpha_level += .1
    ax.plot(crime_per_1000['date'][crime_per_1000.community_area_name == community], 
            crime_per_1000['c_per_1000residents'][crime_per_1000.community_area_name == community],
            alpha = alpha_level, label = community+'(B)', color = 'b')

ax.set_xlabel('Date')
ax.set_ylabel('Crime per 1000 Residents per Month')
ax.set_title('Crimes Rates per 1,0000 Residents for Communities with Top 10 (T) and Bottom 10 (B) Crime Rates')
ax.legend()

fig.savefig('crime_rates_per_month_by_community.png', transparent = False, dpi = 80, bbox_inches = 'tight')
```

**Check how crime is affected by time of day.**

```python
arrests_by_community = crimes[['date', 'arrest', 'community_area_name', 'violent']]
crimes_by_hour_per_community = arrests_by_community.groupby([arrests_by_community['community_area_name'],
                                                             arrests_by_community['date'].dt.hour]).count()
del crimes_by_hour_per_community['date']
del crimes_by_hour_per_community['violent']
crimes_by_hour_per_community = crimes_by_hour_per_community.reset_index()
crimes_by_hour_per_community.columns = ['community_area_name', 'hour', 'crimes']
arrests_by_hour_per_community = arrests_by_community.groupby([arrests_by_community['community_area_name'],
                                                            arrests_by_community['date'].dt.hour]).sum().reset_index()
arrests_by_hour_per_community.columns = ['community_area_name', 'hour', 'arrests', 'violent_crimes']
arrest_percent_per_community_hour = crimes_by_hour_per_community.merge(arrests_by_hour_per_community[['arrests', 
                                        'violent_crimes']], left_index = True, right_index = True, how = 'left')
```

```python
crime_per_1000_hour = arrest_percent_per_community_hour.merge(socioeconomic[['community_area_name', 
                                            '2010_population']], on = 'community_area_name', how = 'left')
```

```python
crime_per_1000_hour['c_per_1000residents'] = crime_per_1000_hour['crimes'] / (crime_per_1000_hour['2010_population']/1000)
crime_per_1000_hour['v_per_1000residents'] = crime_per_1000_hour['violent_crimes'] / (crime_per_1000_hour['2010_population']/1000)
```

```python
crime_per_1000_hour.head()
```

```python
fig, ax = plt.subplots(figsize = (12,10))
alpha_level = 1

for community in top_10_crime:
    ax.plot(crime_per_1000_hour['hour'][crime_per_1000_hour.community_area_name == community], 
            crime_per_1000_hour['c_per_1000residents'][crime_per_1000_hour.community_area_name == community],
            alpha = alpha_level, label = community+'(T)', color = 'r')
    alpha_level -= .1

for community in bottom_10_crime:
    alpha_level += .1
    ax.plot(crime_per_1000_hour['hour'][crime_per_1000_hour.community_area_name == community], 
            crime_per_1000_hour['c_per_1000residents'][crime_per_1000_hour.community_area_name == community],
            alpha = alpha_level, label = community+'(B)', color = 'b')

plt.xlim((0,31))
plt.xticks(np.arange(24))
ax.set_xlabel('Hour')
ax.set_ylabel('Crime per 1000 Residents per Hour of Day (In Total - 3 Years)')
ax.set_title('Crime per 1000 Residents per Hour of Day for Communities with Top 10 (T) and Bottom 10 (B) Crime Rates')
ax.legend()

fig.savefig('crime_rates_per_hour_by_community.png', transparent = False, dpi = 80, bbox_inches = 'tight')
```

**Check for correlation in between various socioeconomic factors and crime.**

```python
crime_correlation = crime_per_1000_total[['income_per_capita', 'percent_over25_without_diploma',
                                          'percent_households_below_poverty', 'percent_over16_unemployed',
                                          'mean_yearly_crime_rate_per1000', 'mean_yearly_violent_rate_per1000']]
```

```python
crime_correlation.columns = ['income', 'HS_diploma', 'poverty', 'unemployment', 'crime_rate', 'violent_crime_rate']
```

```python
from pandas.tools.plotting import table
```

```python
corr_table = crime_correlation.corr(method = 'pearson')

fig, ax = plt.subplots(figsize = (12, 2)) 
ax.xaxis.set_visible(False)  
ax.yaxis.set_visible(False)  
ax.set_frame_on(False)
tab = table(ax, corr_table, loc='upper right')
tab.scale(1.2, 1.2)

plt.savefig('corr_table.png', transparent = False, dpi = 300, bbox_inches = 'tight')
```
